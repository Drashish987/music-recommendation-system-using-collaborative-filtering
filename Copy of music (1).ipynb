# ***MAIN PROJECT OF MUSIC RECOMMENDATON SYSTEM using colaborating fltering where you will give the number of id as any arbitary value and the output will be the name of artists***

```python
from pathlib import Path

import scipy
import pandas as pd


def load_user_artists(user_artists_file: Path) -> scipy.sparse.csr_matrix:
    """Load the user artists file and return a user-artists matrix in csr
    fromat.
    """
    user_artists = pd.read_csv(user_artists_file, sep="\t")
    user_artists.set_index(["userID", "artistID"], inplace=True)
    coo = scipy.sparse.coo_matrix(
        (
            user_artists.weight.astype(float),
            (
                user_artists.index.get_level_values(0),
                user_artists.index.get_level_values(1),
            ),
        )
    )
    return coo.tocsr()


class ArtistRetriever:
    """The ArtistRetriever class gets the artist name from the artist ID."""

    def __init__(self):
        self._artists_df = None

    def get_artist_name_from_id(self, artist_id: int) -> str:
        """Return the artist name from the artist ID."""
        return self._artists_df.loc[artist_id, "name"]

    def load_artists(self, artists_file: Path) -> None:
        """Load the artists file and stores it as a Pandas dataframe in a
        private attribute.
        """
        artists_df = pd.read_csv(artists_file, sep="\t")
        artists_df = artists_df.set_index("id")
        self._artists_df = artists_df


if __name__ == "__main__":
    # user_artists_matrix = load_user_artists(
    #     Path("../lastfmdata/user_artists.dat")
    # )
    # print(user_artists_matrix)

    artist_retriever = ArtistRetriever()
    artist_retriever.load_artists(Path("../lastfmdata/artists.dat"))
    artist = artist_retriever.get_artist_name_from_id(9)
    print(artist)
```

# ***MACHINE LEARNING MODEL TRAININGS AND APPLYING ALTERNATING LEAST SQUARE MATRIX FOR THE RECOMMENDER SYSTEMS WHERE WE DIVIDE THE MATRIX INTO TWO HALFS i.e USER MATRIX AND ARTIST MATRIX AND MULTIPLY USING ALS TO FIND THE RATING OF UNKNOWN RATINGS OF A PARTICULAR MUSICS***

```python
from typing import Tuple, List # Import Tuple and List from typing module
import sys
import os

# Get the absolute path of the current notebook
notebook_path = os.path.abspath('')

# Get the directory containing the notebook (one level up)
project_dir = os.path.dirname(notebook_path)

# Navigate to the root of the cloned project
project_root = os.path.join(project_dir, 'musiccollaborativefiltering')

# Add the root of the project (and subfolders like 'musiccollaborativefiltering')
# to the Python path, allowing you to import from any location within it.
if project_root not in sys.path:
    sys.path.insert(0, project_root)


# Assuming the project has the following structure:
# /project_root
#   /musiccollaborativefiltering
#     __init__.py
#     data.py
#     ...
#    your_notebook.ipynb

from pathlib import Path

import scipy
import pandas as pd
import implicit
import data


class ImplicitRecommender:
    """The ImplicitRecommender class computes recommendations for a given user
    using the implicit library.
    Attributes:
        - artist_retriever: an ArtistRetriever instance
        - implicit_model: an implicit model
    """

    def __init__(
        self,
        artist_retriever: ArtistRetriever,
        implicit_model: implicit.recommender_base.RecommenderBase,
    ):
        self.artist_retriever = artist_retriever
        self.implicit_model = implicit_model

    def fit(self, user_artists_matrix: scipy.sparse.csr_matrix) -> None:
        """Fit the model to the user artists matrix."""
        self.implicit_model.fit(user_artists_matrix)

    def recommend(
        self,
        user_id: int,
        user_artists_matrix: scipy.sparse.csr_matrix,
        n: int = 10,
    ) -> Tuple[List[str], List[float]]: # Use Tuple after importing it
        """Return the top n recommendations for the given user."""
        artist_ids, scores = self.implicit_model.recommend(
            user_id, user_artists_matrix[user_id], N=n  # Indexing by user_id
        )
        artists = [
            self.artist_retriever.get_artist_name_from_id(artist_id)
            for artist_id in artist_ids
        ]
        return artists, scores


if __name__ == "__main__":
    # load user artists matrix
    user_artists = load_user_artists(Path("../lastfmdata/user_artists.dat"))
    # instantiate artist retriever
    artist_retriever = ArtistRetriever()
    artist_retriever.load_artists(Path("../lastfmdata/artists.dat"))
    # instantiate ALS using implicit
    implict_model = implicit.als.AlternatingLeastSquares(
        factors=50, iterations=100, regularization=0.01
    )
    # instantiate recommender, fit, and recommend
    recommender = ImplicitRecommender(artist_retriever, implict_model)
    recommender.fit(user_artists)
    artists, scores = recommender.recommend(2, user_artists, n=78)
    # print results
    for artist, score in zip(artists, scores):
        print(f"{artist}: {score}")
```

# ***SINGULAR VALUE DECOMPOSITIONS (SVD) AND ROOT MEAN SQUARED ERRORS ***

```python
import pandas as pd
from surprise import Dataset, Reader
from surprise import SVD
from surprise.model_selection import train_test_split
from surprise import accuracy

# Load the dataset - replace 'your_dataset.csv' with actual dataset
# Sample data structure: user_id, item_id, rating
data = pd.read_csv('/content/musiccollaborativefiltering/lastfmdata/user_artists.dat', sep='\t')  # Upload your own collaborative filtering dataset here
reader = Reader(rating_scale=(1, 5))

# Prepare the dataset for Surprise library
# Use actual column names from your dataset
surprise_data = Dataset.load_from_df(data[['userID', 'artistID', 'weight']], reader)
trainset, testset = train_test_split(surprise_data, test_size=0.2)

# Model training
model = SVD()
model.fit(trainset)

# Model evaluation
predictions = model.test(testset)
accuracy.rmse(predictions)
```

```python
# Define a recommendation function
def get_recommendations(user_id, num_recommendations=5):
    # Assuming we have a list of all item_ids
    item_ids = data['item_id'].unique()
    predictions = []

    # Predict rating for each item and store it
    for item_id in item_ids:
        predictions.append((item_id, model.predict(user_id, item_id).est))

    # Sort by estimated rating
    recommendations = sorted(predictions, key=lambda x: x[1], reverse=True)[:num_recommendations]
    return [item[0] for item in recommendations]
```

# ***Data Preparation: Loading Data into Surprise Format: Train-Test Split: Model Initialization and Training: Making Predictions: Evaluating Accuracy: Matrix Factorization Explanation:


 ***

```python
import numpy as np
from surprise import SVD, Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import accuracy
import pandas as pd

# Sample Data: User-Item Ratings (replace this with your actual data)
data_dict = {
    'user_id': ['A', 'A', 'B', 'B', 'C', 'C', 'D', 'D'],
    'item_id': ['Song1', 'Song2', 'Song1', 'Song3', 'Song2', 'Song3', 'Song1', 'Song2'],
    'rating': [5, 3, 4, 2, 3, 5, 4, 3]
}
df = pd.DataFrame(data_dict)

# Load data into Surprise dataset format
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df[['user_id', 'item_id', 'rating']], reader)

# Train-test split
trainset, testset = train_test_split(data, test_size=0.25)

# Initialize SVD algorithm
algo = SVD()

# Train the algorithm on the trainset, and predict ratings for the testset
algo.fit(trainset)
predictions = algo.test(testset)

# Evaluate accuracy (RMSE)
rmse = accuracy.rmse(predictions)

# Display predictions for each user-item pair in test set
print("\nPredictions:")
for prediction in predictions:
    print(f"User: {prediction.uid}, Item: {prediction.iid}, True Rating: {prediction.r_ui:.2f}, Predicted Rating: {prediction.est:.2f}")

# Matrix Factorization Explanation (for visualization)
# Display the decomposed matrices (simulated here)
print("\nUser and Item Matrix Factorization (Latent Factors):")
user_factors = algo.pu  # User latent factors
item_factors = algo.qi  # Item latent factors
print(f"User Factors (Sample):\n{user_factors[:2]}")
print(f"Item Factors (Sample):\n{item_factors[:2]}")
```

# ***Setup: Data Preparation: ALS Model Training: Latent Factor Extraction:  Visualization:

  ***

```python
from pyspark.ml.recommendation import ALS
from pyspark.sql import SparkSession
from pyspark.sql import Row
import matplotlib.pyplot as plt
import numpy as np

# Initialize Spark session
spark = SparkSession.builder.appName("ALSExample").getOrCreate()

# Sample data: User, Item, and Ratings
data = [
    Row(userId=0, itemId=1, rating=4.0),
    Row(userId=0, itemId=3, rating=2.0),
    Row(userId=1, itemId=1, rating=3.0),
    Row(userId=1, itemId=2, rating=4.0),
    Row(userId=2, itemId=0, rating=2.0),
    Row(userId=2, itemId=3, rating=5.0),
]

# Create DataFrame
df = spark.createDataFrame(data)

# ALS Model
als = ALS(maxIter=10, regParam=0.1, userCol="userId", itemCol="itemId", ratingCol="rating", rank=2)
model = als.fit(df)

# Extract latent factors
user_factors = model.userFactors.select("features").collect()
item_factors = model.itemFactors.select("features").collect()

# Convert to NumPy arrays for visualization
user_factors_np = np.array([row["features"] for row in user_factors])
item_factors_np = np.array([row["features"] for row in item_factors])

# Plot User and Item Latent Factors
plt.figure(figsize=(10, 5))

# User factors
plt.subplot(1, 2, 1)
plt.title("User Latent Factors")
plt.xlabel("Latent Factor 1")
plt.ylabel("Latent Factor 2")
plt.scatter(user_factors_np[:, 0], user_factors_np[:, 1], color='blue', label='Users')
plt.legend()

# Item factors
plt.subplot(1, 2, 2)
plt.title("Item Latent Factors")
plt.xlabel("Latent Factor 1")
plt.ylabel("Latent Factor 2")
plt.scatter(item_factors_np[:, 0], item_factors_np[:, 1], color='green', label='Items')
plt.legend()

plt.show()

# Stop Spark session
spark.stop()
```
